```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
dataset = pd.DataFrame(data=data['data'], columns=data['feature_names'])
dataset
```

```python
from sklearn.model_selection import train_test_split

X = dataset.copy()
y = data['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python
clf.get_params()
```

```python
predictions = clf.predict(X_test)
predictions
```

```python
# tree that doesn't have stopping criteria (max_depth=None)
clf.predict_proba(X_test)[:20]
```

```python
clf = DecisionTreeClassifier(max_depth=4)
clf = clf.fit(X_train, y_train)
```

```python
clf.get_params()
```

```python
predictions = clf.predict(X_test)
predictions
```

```python
# tree that has stopping criteria (max_depth=4)
clf.predict_proba(X_test)[:20]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
from sklearn.metrics import precision_score
precision_score(y_test, predictions)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test, predictions)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions, labels=[0,1])
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions, target_names=['malignant', 'benign']))
```

```python
# ccp_alpha: non-negative float, default=0.0
# complexity parameter used for Minimal Cost-Complexity Pruning. The subtree with the largest cost complexity that is smaller
# than ccp_alpha will be chosen
```

```python
clf = DecisionTreeClassifier(ccp_alpha=0.01)
clf = clf.fit(X_train, y_train)
```

```python
predictions = clf.predict(X_test)
```

```python
from sklearn import tree
from matplotlib import pyplot as plt

fig = plt.figure(figsize=(25, 20))
_ = tree.plot_tree(clf,
                   feature_names=feature_names,
                   class_names={0: 'Malignant', 1: 'Benign'},
                   filled=True,
                   fontsize=12)
```

```python
accuracy_score(y_test, predictions)
```
