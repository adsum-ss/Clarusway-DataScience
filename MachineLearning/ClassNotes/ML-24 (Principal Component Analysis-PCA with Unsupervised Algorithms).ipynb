## PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)  # default=None
```

```python
pca.get_params()
```

```python
# n_components ===> datadaki feature ve gözlem sayısından hangisi minimum ise onu dikkate alır, 
# yani 10 gözlem 30 feature ya da 10 feature 30 gözlem varsa 10 component oluşturur
```

```python
principal_components = pca.fit_transform(scaled_X)
```

```python
component_df = pd.DataFrame(data = principal_components, columns = ["first_component", 
                                                                    "second_component"])
component_df
```

## K-Means Algorithm

```python
from pyclustertend import hopkins
```

```python
hopkins(component_df, component_df.shape[0])
```

```python
from sklearn.cluster import KMeans

ssd = []

K = range(2,10)

for k in K:
    model = KMeans(n_clusters =k, random_state=42)
    model.fit(component_df)
    ssd.append(model.inertia_)
```

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error") 
plt.title("elbow method")
```

```python
ssd
```

```python
pd.Series(ssd).diff()
```

```python
df_diff =pd.DataFrame(-pd.Series(ssd).diff()).rename(index = lambda x : x+1)
df_diff
```

```python
df_diff.plot(kind='bar')
```

```python
from yellowbrick.cluster import KElbowVisualizer

model_ = KMeans(random_state=42)
visualizer = KElbowVisualizer(model_, k=(2,9))

visualizer.fit(component_df)        # Fit the data to the visualizer
visualizer.show();
```

```python
from sklearn.metrics import silhouette_score

range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(component_df)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(component_df, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```

```python
sns.scatterplot(x = component_df.first_component, y= component_df.second_component, alpha=0.7)
```

```python
# 3 ve 4 küme sayıları için kanser evreleri olarak düşünülebilir
# ama aldığımız target label'e göre iki küme olduğunu biliyoruz (kanser / kanser değil)
```

```python
model = KMeans(n_clusters =2, random_state=42)
clusters = model.fit_predict(component_df)
clusters
```

```python
ct = pd.crosstab(y, clusters)
ct
# model toplam 53 hata yapmış
```

## Interpreting PCA results

```python
scaled_df = pd.DataFrame(scaled_X, columns=X.columns)
scaled_df.head()
```

```python
combined_df = pd.concat([component_df, scaled_df], axis =1)
correlation = combined_df.corr()
correlation.head()
```

```python
fig, ax = plt.subplots(figsize=(20, 7))
sns.heatmap(correlation.iloc[:2,2:], cmap='YlGnBu', linewidths=.7, annot=True, fmt='.2f')
plt.show()
# component'ler ile içindeki feature'ların korelasyonları
# yorumlarken 0.5 ve üzeri olan feature'ları dikkate alıyoruz, ancak büyük olan feature daha önemlidir yorumunu yapamayız
# (bu yorumu sadece feature importance ile yapabiliriz - target label elimizde olduğu zaman)
# sadece 1.component için şu feature'ın artması şu feature'ın da artması sonucunu doğurur vb yorumlar yapılabilir
```

```python
# domain araştırması sonucunda 'mean fractal dimension' feature'nın hem kanser olan hem olmayan hastalarda bazı kanser türleri
# için dengesiz sonuçlar verdiğini, hastalık teşhisinde çok etkili olmadığını öğrendik; bu yüzden onu kullanmayacağız

# 1.component ===> hastalığın teşhisi için önemli olan feature'lar
# 2.component ===> hasta olmayanlar için önemli feature'lar yorumunu yapabiliriz
```
