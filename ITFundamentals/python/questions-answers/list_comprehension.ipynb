```python
#  Create a list of strings based on a list of numbers
#  The rules:

#  If the number is a multiple of five and odd, the string should be 'five odd'
#  If the number is a multiple of five and even, the string should be 'five even'
#  If the number is odd, the string is 'odd'
#  If the number is even, the string is 'even'

#  we can solve this guestion using loop easily it is okey. But can we solve it using list comprehension ? 
#  given list =numbers = [1, 3, 4, 6, 81, 80, 100, 95] 
#  expected output = ['odd', 'odd', 'even', 'even', 'odd', 'five even', 'five even', 'five odd']
```

```python
numbers = [1, 3, 4, 6, 81, 80, 100, 95]

num_str = ["five_even" if n%5==0 and n%2==0 else "five_odd" if n%5==0 else "even" if n%2==0 else "odd" for n in numbers]

print(num_str)
```

```python
numbers = [1, 3, 4, 6, 81, 80, 100, 95]

num_str = [['odd', 'even', 'five_odd', 'five_even'][2*(n%5==0)+(n%2==0)] for n in numbers]

print(num_str)
```
