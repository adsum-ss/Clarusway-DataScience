```python
def JosephusProblem(n , k):
  ans = 0

  for size in range(2, n+1):
    ans = (ans + k) % size

  return ans

print(JosephusProblem(16,9))
```

```python
def who_goes_free(n, k):
    pos = 0
    for i in range(1, n + 1):
        pos += k
        pos %= i
    return pos

who_goes_free(16,9)
```

```python
def who_goes_free(people, step):

  people = list(range(people))

  while len(people) != 1:
    if len(people) < step:
      if step % len(people) == 0:
        people.pop(-1)
      else:
        k = step % len(people)
        people.pop(k-1)
        people = (people[k-1:] + people[0:k-1])
    else:
      people.pop(step-1)
      people = people[step-1:] + people[0:step-1]

  return people[0]

who_goes_free(16,9)
```
