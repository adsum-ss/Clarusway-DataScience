```python
# If you had deposited a coin on the cryptocurrency exchange that brought 7% fixed profit daily for a week, 
# how much would your $ 1000 reach at the end of the 7th day?

capital = 1000
profit = 0.07
term = 7

account = capital * (1 + profit)**term
print(account)
```
