### 2
```
A number is said to be Harshad if it's exactly divisible by the sum of its digits. Create a function that determines whether a number is a Harshad or not.

is_harshad(75) ➞ False
#7 + 5 = 12
#75 is not exactly divisible by 12
 
is_harshad(171) ➞ True
#1 + 7 + 1 = 9
#9 exactly divides 171
 
is_harshad(481) ➞ True

is_harshad(89) ➞ False

is_harshad(516) ➞ True

is_harshad(200) ➞ True
```

````python
def is_harshad(num):
  digits = sum([int(i) for i in list(str(num))])
  harshad = num % digits == 0
  
  return harshad 

is_harshad(75)
````

### 3
```
"Loves me, loves me not" is a traditional game in which a person plucks off all the petals of a flower one by one, saying the phrase "Loves me" and "Loves me not" when determining whether the one that they love, loves them back.

Given a number of petals, return a string which repeats the phrases "Loves me" and "Loves me not" for every alternating petal, and return the last phrase in all caps. Remember to put a comma and space between phrases.
```

````python
def game(petals):
 
  phrase1 = "Loves me"
  phrase2 = "Loves me not"
  string = ""

  for i in range(petals-1):
    if i % 2 == 0:
      string += phrase1 + ", "
    else:
      string += phrase2 + ", "

  if petals % 2 == 0:
    string += "LOVES ME NOT"
  else:
    string += "LOVES ME"

  return string

game(6)
````

### 4
```
Create a function that takes an integer n and returns the factorial of factorials. 

See below examples for a better understanding:

fact_of_fact(4) ➞ 288
#4! * 3! * 2! * 1! = 288

fact_of_fact(5) ➞ 34560

fact_of_fact(6) ➞ 24883200
```

````python
def fact_of_fact(num):
  
  total = 1
  fact = 1
  for n in range(1, num+1):
    total *= n
    fact *= total

  return fact  

fact_of_fact(6)
````

### 5
```
Given a list of words in the singular form, return a set of those words in the plural form if they appear more than once in the list.

pluralize(["cow", "pig", "cow", "cow"]) ➞ { "cows", "pig" }

pluralize(["table", "table", "table"]) ➞ { "tables" }

pluralize(["chair", "pencil", "arm"]) ➞ { "chair", "pencil", "arm" }
```

````python
def pluralize(words):
  words2 = set()
  for i in words:
    x = words.count(i)
    if x > 1:
      i += "s"
      words2.add(i)
    else:
      words2.add(i)

  return words2

pluralize(["cow", "pig", "cow", "cow"])
````

### 6
```
Mubashir needs your help to construct a building which will be a pile of n cubes. The cube at the bottom will have a volume of n^3, the cube above will have volume of (n-1)^3 and so on until the top which will have a volume of 1^3.

Given the total volume m of the building, can you find the number of cubes n required for the building?

In other words, you have to return an integer n such that:
n^3 + (n-1)^3 + ... + 1^3 == m
Return None if there is no such number.

pile_of_cubes(1071225) ➞ 45

pile_of_cubes(4183059834009) ➞ 2022

pile_of_cubes(16) ➞ None
```

````python
def pile_of_cubes(num):
  m = 0
  cubes = []
  for i in range(1, num):
    m += i ** 3
    cubes.append(i)
    if m == num:
      return i
      break;
    elif m > num:
      return 'None'

pile_of_cubes(35)
````

### 8
```
Create a function that takes a number num and returns each place value in the number.

num_split(39) ➞ [30, 9]

num_split(-434) ➞ [-400, -30, -4]

num_split(100) ➞ [100, 0, 0]
```

````python
def num_split(num):
  liste = []
  negative = False
  if num.startswith("-"):
    num = num.lstrip("-")
    negative = True
  num = "".join(reversed(num))
  for i in range(len(num)):
    if negative:
      liste.append(-int(num[i]) * (10 ** i))
    else:
      liste.append(int(num[i]) * (10 ** i))
   
  return liste[::-1]

num = input("Enter a number: ")
num_split(num)
````

### 10
```
Write a function that groups a string into parentheses cluster. Each cluster should be balanced.
split("()()()") ➞ ["()", "()", "()"]

split("((()))") ➞ ["((()))"]

split("((()))(())()()(()())") ➞ ["((()))", "(())", "()", "()", "(()())"]

split("((())())(()(()()))") ➞ ["((())())", "(()(()()))"]
```

````python
text = "((())())(()(()()))"
left, right = 0, 0
phase = ''
result = []
for i in text:
    if i == '(':
        left += 1
        phase += i
    elif i == ')':
        right += 1
        phase += i
    if left == right:
        result.append(phase)
        phase = ''
print(result)
````

```
Examples lcm([1, 2, 3, 4, 5, 6, 7, 8, 9]) ➞ 2520

lcm([5]) ➞ 5

lcm([5, 7, 11]) ➞ 385

lcm([5, 7, 11, 35, 55, 77]) ➞ 385
```

````python
def lcm(nums):

    counter = 1
    dividing = list(map(lambda x: (max(nums) * counter) % x == 0, nums))

    while False in dividing:
    
      counter += 1
      dividing = list(map(lambda x: (max(nums) * counter) % x == 0, nums))
  
    return max(nums) * counter
  
lcm([5, 7, 11, 35, 55, 77])
````
