**Assignment-009/8 (Letters Count)**

**Task:**

Count the number of each letter in a sentence.

* The department you work for undertook a project construction that makes word / text analysis. 
* You are asked to calculate the number of letters or any chars in the sentences entered under this project.

  Write a Python program that;
    1. takes a sentence from the user,
    2. counts the number of each letter of the sentence,
    3. collects the letters/chars as a key and the counted numbers as a value in a dictionary.

**Examples**

*Sample inputs*	
hippo runs to us!

*Outputs*

{'s': 2, 'r': 1, 't': 1, 'h': 1, 'n': 1, 
'i': 1, 'u': 2, 'o': 2, 'p': 2, ' ': 3, '!': 1}

Do not use try - except block.

```python
sentence = input("Enter a sentence: ")

chars = list(sentence)
repetitions = list(map(lambda char: chars.count(char), chars))

char_count = dict(zip(chars, repetitions))

print(char_count)
```

```python
def char_count(sentence):
  chars = list(sentence)
  repetitions = [chars.count(char) for char in chars]

  char_count = dict(zip(chars, repetitions))
  return char_count

sentence = input("Enter a sentence: ")
print(char_count(sentence))
```
